## Overpass API

```python
# URL of the Overpass API
overpass_url = "http://overpass-api.de/api/interpreter"

# Overpass query to get detailed information about restaurants in the Paris area
overpass_query = """
[out:json];
area[name="Paris"]->.searchArea;
node["amenity"="restaurant"](area.searchArea);
out body;
"""

# Execute the HTTP request
response = requests.get(overpass_url, params={'data': overpass_query})
data = response.json()

# Extract relevant information (name, latitude, longitude, address, opening hours, etc.)
restaurants = []
for element in data['elements']:
    if element['type'] == 'node':
        # Retrieve name and geographic coordinates
        name = element.get('tags', {}).get('name', 'Unnamed Restaurant')
        lat = element.get('lat')
        lon = element.get('lon')
        
        # Retrieve additional information if available
        address = element.get('tags', {}).get('addr:full')  # Full address if available
        street = element.get('tags', {}).get('addr:street')
        house_number = element.get('tags', {}).get('addr:housenumber')
        postcode = element.get('tags', {}).get('addr:postcode')
        city = element.get('tags', {}).get('addr:city')
        cuisine = element.get('tags', {}).get('cuisine')  # Type of cuisine
        website = element.get('tags', {}).get('website')
        phone = element.get('tags', {}).get('contact:phone')
        opening_hours = element.get('tags', {}).get('opening_hours')
        
        # Append the data to the list
        restaurants.append({
            'name': name,
            'lat': lat,
            'lon': lon,
            'address': address,
            'street': street,
            'house_number': house_number,
            'postcode': postcode,
            'city': city,
            'cuisine': cuisine,
            'website': website,
            'phone': phone,
            'opening_hours': opening_hours
        })

# Convert to a pandas DataFrame for easy manipulation
df_restaurants = pd.DataFrame(restaurants)

# Display the first few results
print(df_restaurants.head())

# Save the data to a CSV file
df_restaurants.to_csv('restaurants_paris_osm.csv', index=False)
```

## WEB SCRAPPING "https://www.linternaute.com" - CLEAN & EDA

```python
#Scrapping population in France through the site "https://www.linternaute.com
# URL to scrap
base_url = "https://www.linternaute.com/ville/classement/villes/population?page="


data = []

# Number of pages to scrap
num_pages = 699

for page in range(1, num_pages + 1):
    url = base_url + str(page)
    print(f"Scraping page {page}...")
    
    # Request
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Erreur lors de l'accès à la page {page}")
        continue
    
    # Analyse the page with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Getting the table with the info
    table = soup.find("table", {"class": "odTable"})
    if not table:
        print(f"Tableau non trouvé à la page {page}")
        continue
    
    # Getting the info
    rows = table.find("tbody").find_all("tr")
    for row in rows:
        cells = row.find_all("td")
        
        if len(cells) >= 3:
            rank = cells[0].text.strip()
            city = cells[1].text.strip()
            population = cells[2].text.strip()
            
            # Append the info on the list
            data.append({"Rank": rank, "City": city, "Population": population})
    
    
    time.sleep(1)


df = pd.DataFrame(data)

# Saving the data
df.to_csv("population_france.csv", index=False, encoding="utf-8")
print("Données sauvegardées dans population_france.csv")
```

```python
#Plotting population by arrondissement
df['Population'] = df['Population'].astype(str).str.replace("habitants", "").str.replace("\xa0", "").str.replace(" ", "").astype(int)

# Filter for the top 10 most populated cities
df_top10 = df.head(10)

# Create a bar plot 
plt.figure(figsize=(10, 6))
bars = plt.bar(df_top10['City'], df_top10['Population'], color=palette)
plt.title('Top 10 Most Populous Cities in France')
plt.xlabel('City')
plt.ylabel('Population')
plt.xticks(rotation=45)
# Add population numbers above each bar
for bar in bars:
    yval = bar.get_height()  # Get the height of each bar (population value)
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:,}', ha='center', va='bottom')  # Display the population value above the bar
plt.tight_layout()
plt.show()
```

## WEB SCRAPPING WIKIPEDIA

```python
# URL of the Wikipedia page
url = "https://fr.wikipedia.org/wiki/Arrondissements_de_Paris"

# Send a GET request to fetch the page content
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find all tables with the class "wikitable" and select the second table (index 1)
tables = soup.find_all("table", {"class": "wikitable"})
table = tables[1]  # Table with index 1 is the one with arrondissement data

# Lists to hold extracted data
arrondissements = []
superficies = []
populations_2020 = []
densities_2021 = []

# Loop through each row in the table body, skipping the header row
for row in table.find("tbody").find_all("tr")[1:]:
    cells = row.find_all("td")
    
    # Ensure there are enough cells in the row to extract required data
    if len(cells) >= 13:  # The table structure requires checking for at least 13 columns
        arrondissement = cells[0].get_text(strip=True)
        superficie = cells[1].get_text(strip=True)
        population_2020 = cells[7].get_text(strip=True)  # Corrected to the 8th cell (index 7)
        density_2021 = cells[13].get_text(strip=True)    # Corrected to the 13th cell (index 12)
        
        # Append data to lists
        arrondissements.append(arrondissement)
        superficies.append(superficie)
        populations_2020.append(population_2020.replace('\xa0', ''))
        densities_2021.append(density_2021.replace('\xa0', ''))

# Create a DataFrame from the lists
df_paris_data = pd.DataFrame({
    "Arrondissement": arrondissements,
    "Superficie (ha)": superficies,
    "Population (2020)": populations_2020,
    "Density (2021, hab/km²)": densities_2021
})

# Display the DataFrame
print(df_paris_data)
# Display the resulting DataFrame
df_paris_data.to_csv('population_paris_2021.csv')
```

```python
#with keywords in french
# Load the file containing the reviews
df = pd.read_csv('FINAL_kids_friendly_places_paris_with_reviews.csv')

# Extended list of keywords to identify kids-friendly reviews
keywords_for_kids = [
     "kids menu", "children", "kid-friendly", "for children", "allergen-free dishes", "child-friendly", "family", "children's games", 
    "kids welcome", "family-oriented", "baby", "family", "family-friendly environment", "play area", "children's activity", "games", 
    "family with children", "kids", "kid", "children", "childrens"
]

# Function to analyze the sentiment of a review
def analyze_sentiment(review_text):
    blob = TextBlob(review_text)
    return blob.sentiment.polarity  # Polarity ranges from -1 (negative) to +1 (positive)

# List to store kids-friendly sentiment scores with weighting
kids_friendly_sentiments = []

# Analyze each review
for index, row in df.iterrows():
    review_text = row['reviews']  # Reviews for a place, as text
    rating = row.get('rating', '0')  # Place rating, default to '0' if missing
    
    # Convert rating to a numeric value if possible
    try:
        rating = float(rating)
    except ValueError:
        rating = 0  # Set to 0 if conversion fails

    if pd.notna(review_text):  # Check that the review is not empty
        # Split into sentences to capture kids-friendly mentions
        sentences = review_text.split('.')
        relevant_reviews = [sentence for sentence in sentences if any(keyword in sentence.lower() for keyword in keywords_for_kids)]
        
        # Analyze the sentiment of relevant sentences
        if relevant_reviews:
            sentiment_scores = [analyze_sentiment(sentence) for sentence in relevant_reviews]
            avg_kids_friendly_sentiment = sum(sentiment_scores) / len(sentiment_scores)
        else:
            avg_kids_friendly_sentiment = 0  # If no kids-friendly review, sentiment is neutral
    else:
        avg_kids_friendly_sentiment = 0  # If no review, sentiment is neutral
    
    # Calculate a weighted kids-friendly sentiment score based on the rating
    weighted_sentiment = avg_kids_friendly_sentiment * (rating / 5)  # Scale of 0 to 5
    kids_friendly_sentiments.append(weighted_sentiment)

# Add the weighted kids-friendly sentiment scores to the DataFrame
df['kids_friendly_weighted_sentiment'] = kids_friendly_sentiments

# Save the results to a CSV and Excel file
df.to_csv('Sentiment_final_weighted.csv', index=False)
df.to_excel('Sentiment_final_weighted.xlsx', index=False)

print("The enriched data with the weighted kids-friendly sentiment analysis has been saved.")
```

## Functions definition for ML - NLP

```python
# To use POS with lemmatization we need to create a function

def get_wordnet_pos(word):
  tag = nltk.pos_tag([word], lang="eng")[0][1][0]
  tag_dict = {"R": wordnet.ADV,
              "N": wordnet.NOUN,
              "V": wordnet.VERB,
              "J": wordnet.ADJ}

  return tag_dict.get(tag, wordnet.NOUN)

def tokenize(row):
  tokens = word_tokenize(row)
  return [word.lower() for word in tokens if word.isalpha()]

def lemmatizer_with_pos(row):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in row]

def remove_sw(row):
  return [word for word in row if (word not in stopwords.words("english") and len(word)>1)]
```

```python
# Function to detect keywords with negation handling
def is_kids_friendly(review):
    review = str(review).translate(str.maketrans('', '', string.punctuation)).lower()
    for keyword in keywords_for_kids:
        # Search for the keyword in the review
        match = re.search(r'\b' + keyword + r'\b', review)
        if match:
            # Check for negation words before the keyword (up to 3 words before)
            context = review[max(0, match.start() - 20):match.start()]
            if not any(neg_word in context for neg_word in negation_keywords):
                return True
    return False
    
# Mots-clés indiquant un endroit kids-friendly
keywords_for_kids = ["kids menu", "children", "kid-friendly", "for children", "allergen-free dishes", "child-friendly", "family", "children's games", 
    "kids welcome", "family-oriented", "baby", "family", "family-friendly environment", "play area", "children's activity", "games", 
    "family with children", "kids", "kid", "children", "childrens","kids", "children", "family-friendly",
                     "high chair", "play area", "kids menu", "baby friendly","kids friendly","baby area"]
negation_keywords = ["not", "no", "never", "none", "didn't", "doesn't", "without"]
# Ajout de la colonne 'kids_friendly' pour chaque restaurant
data['kids_friendly'] = data['reviews'].apply(is_kids_friendly)
data.to_csv("Gral_df_Scores.csv", index=False)
```
