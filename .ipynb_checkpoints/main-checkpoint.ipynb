```python

#We will start with a simple search to see the current trends on a child-related term, such as "kid-friendly restaurant."

# Starting with pytrends
pytrends = TrendReq(hl='fr-FR', tz=360)

# Request for "restaurant kids friendly"
pytrends.build_payload(kw_list=['restaurant kids friendly'], timeframe='today 12-m', geo='FR')

# Getting trends over the time
trends_kids_friendly = pytrends.interest_over_time()

# Check the got info 
print(trends_kids_friendly.head())
trends_kids_friendly.to_csv("trends_search_restaurants.csv")
# Plotting trends
trends_kids_friendly.plot(figsize=(10,6))
plt.title('Search Trends for "kid-friendly restaurant" in France (Last 12 Months)')
plt.xlabel("Date")
plt.ylabel("Popularity (Index)")
plt.show()
```

Trend Chart for "kid-friendly restaurant": Interest in "kid-friendly restaurant" seems to fluctuate throughout the year, with peaks in May, June, August, and around March.
Searches show sharp declines in December and early summer, likely related to holiday periods or shifts in family habits.

```python
#Cleaning que-faire-a-paris-.xlsx Kid's evenements in Paris Data - Data got it from Kaggle
df_faire= pd.read_csv('que-faire-a-paris-.csv',delimiter=';')
df_faire = df_faire[~df_faire['audience'].isin(['Public adultes',"Public adultes. A partir de 15 ans.","Public jeunes. A partir de 15 ans. Jusqu'à 18 ans.","Public adultes. A partir de 18 ans. Jusqu'à 99 ans.",'Public adultes. A partir de 16 ans.','Public adultes. A partir de 55 ans.','Public jeunes et adultes. A partir de 13 ans.','Public jeunes et adultes. A partir de 14 ans.','Public adultes. A partir de 18 ans.', 'Public adultes.','Public jeunes et adultes. A partir de 15 ans.',"Public adultes. A partir de 15 ans. Jusqu'à 132 ans.", "Public adultes. A partir de 15 ans. Jusqu'à 99 ans.",'Public adultes. A partir de 13 ans.', 'Public jeunes et adultes. A partir de 12 ans.','Public jeunes et adultes.', 'Public jeunes et adultes. A partir de 16 ans.'])]
df_faire["Type d'accès"] = df_faire["Type d'accès"].fillna('conseillée')
df_faire["Type de prix"] = df_faire["Type de prix"].fillna('payant')
df_faire["Coordonnées géographiques"] = df_faire["Coordonnées géographiques"].fillna('Unknow')
df_faire=df_faire.drop(columns=['Occurrences','Mots clés','title_event','childrens','Accès PMR','Accès mal voyant','Accès mal entendant','Accès mal entendant','Transport','Url de contact','Téléphone de contact','Email de contact','URL Facebook associée','URL Twitter associée','URL de réservation',"URL de l'image",'Détail du prix','Date de mise à jour','URL de réservation - Texte', "Texte alternatif de l'image","Crédit de l'image",'Image de couverture','Programmes','En ligne - address_text','En ligne - address_url', 'En ligne - address_url_text'])
df_faire=df_faire.dropna(subset=['Date de début','Date de fin','Description de la date','Nom du lieu','Adresse du lieu','Code postal','Ville'], how='all')
df_faire=df_faire.dropna(subset=['Nom du lieu','Adresse du lieu','Code postal','Ville'], how='all')
df_faire["Nom du lieu"] = df_faire["Nom du lieu"].fillna(df_faire["Code postal"])
df_faire["Description"] = df_faire["Description"].fillna(df_faire["Chapeau"])
df_faire["Chapeau"] = df_faire["Chapeau"].fillna(df_faire["Titre"])
df_faire["Description de la date"] = df_faire["Type de prix"].fillna('Unknow')
#Convert 'Date de début' and 'Date de fin' columns, handling missing or invalid dates with 'coerce'
df_faire['Date de début'] = pd.to_datetime(df_faire['Date de début'], errors='coerce')
df_faire['Date de fin'] = pd.to_datetime(df_faire['Date de fin'], errors='coerce')
#Replacing null values with 'Unknown'
df_faire['Date de début'] = df_faire['Date de début'].apply(lambda x: 'Unknown' if pd.isna(x) else x)
df_faire['Date de fin'] = df_faire['Date de fin'].apply(lambda x: 'Unknown' if pd.isna(x) else x)
#Extract month and year for valid dates only
df_faire['Month_debut'] = df_faire['Date de début'].apply(lambda x: x.month if isinstance(x, pd.Timestamp) else 'Unknown')
df_faire['Year_debut'] = df_faire['Date de début'].apply(lambda x: x.year if isinstance(x, pd.Timestamp) else 'Unknown')
df_faire['Month_fin'] = df_faire['Date de fin'].apply(lambda x: x.month if isinstance(x, pd.Timestamp) else 'Unknown')
df_faire['Year_fin'] = df_faire['Date de fin'].apply(lambda x: x.year if isinstance(x, pd.Timestamp) else 'Unknown')
# Mapping dictionary to convert month numbers to month names
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
    7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
# Apply the mapping to 'Month_fin' and 'Month_debut' columns
df_faire['Month_fin'] = df_faire['Month_fin'].map(month_mapping)
df_faire['Month_debut'] = df_faire['Month_debut'].map(month_mapping)
#dropping unnecesary rows
df_faire['Year_debut'] = df_faire['Year_debut'].replace('Unknown', pd.NA)
df_faire['Year_fin'] = df_faire['Year_fin'].replace('Unknown', pd.NA)
df_faire = df_faire.dropna(subset=['Year_debut', 'Year_fin'])
df_faire['Year_debut'] = df_faire['Year_debut'].astype(int)
df_faire['Year_fin'] = df_faire['Year_fin'].astype(int)
df_faire=df_faire.drop(columns=['Date de début','Date de fin'])
df_faire = df_faire.rename(columns={'ID': 'unique_id', 'Code postal': 'zipcode'})
df_faire['type']='Event'
df_faire['type_id'] = df_faire.apply(lambda row: f"{row['type'][:2].replace(' ', '')}-{row['zipcode']}", axis=1)
df_faire['df_id']='kfevkg75'
df_faire['df_unique_id']=df_faire['df_id']+'_'+(df_faire.index+1).astype(str)
#df_faire.columns = [col.capitalize() for col in df_faire.columns]
df_faire.to_csv('qque-faire-a-paris-.csv',encoding="utf-8")
```

```python
#with keywords in french
# Load the file containing the reviews
df = pd.read_csv('FINAL_kids_friendly_places_paris_with_reviews.csv')

# Extended list of keywords to identify kids-friendly reviews
keywords_for_kids = [
     "kids menu", "children", "kid-friendly", "for children", "allergen-free dishes", "child-friendly", "family", "children's games", 
    "kids welcome", "family-oriented", "baby", "family", "family-friendly environment", "play area", "children's activity", "games", 
    "family with children", "kids", "kid", "children", "childrens"
]


# List to store kids-friendly sentiment scores with weighting
kids_friendly_sentiments = []

# Analyze each review
for index, row in df.iterrows():
    review_text = row['reviews']  # Reviews for a place, as text
    rating = row.get('rating', '0')  # Place rating, default to '0' if missing
    
    # Convert rating to a numeric value if possible
    try:
        rating = float(rating)
    except ValueError:
        rating = 0  # Set to 0 if conversion fails

    if pd.notna(review_text):  # Check that the review is not empty
        # Split into sentences to capture kids-friendly mentions
        sentences = review_text.split('.')
        relevant_reviews = [sentence for sentence in sentences if any(keyword in sentence.lower() for keyword in keywords_for_kids)]
        
        # Analyze the sentiment of relevant sentences
        if relevant_reviews:
            sentiment_scores = [analyze_sentiment(sentence) for sentence in relevant_reviews]
            avg_kids_friendly_sentiment = sum(sentiment_scores) / len(sentiment_scores)
        else:
            avg_kids_friendly_sentiment = 0  # If no kids-friendly review, sentiment is neutral
    else:
        avg_kids_friendly_sentiment = 0  # If no review, sentiment is neutral
    
    # Calculate a weighted kids-friendly sentiment score based on the rating
    weighted_sentiment = avg_kids_friendly_sentiment * (rating / 5)  # Scale of 0 to 5
    kids_friendly_sentiments.append(weighted_sentiment)

# Add the weighted kids-friendly sentiment scores to the DataFrame
df['kids_friendly_weighted_sentiment'] = kids_friendly_sentiments

# Save the results to a CSV and Excel file
df.to_csv('Sentiment_final_weighted.csv', index=False)
df.to_excel('Sentiment_final_weighted.xlsx', index=False)

print("The enriched data with the weighted kids-friendly sentiment analysis has been saved.")
```

```python
# List of keywords to identify "kids-friendly" reviews
keywords_for_kids = [
    "kids menu", "children", "kid-friendly", "family-friendly", "allergen-free", 
    "child-friendly", "family", "play area", "children welcome", "family environment", 
    "baby", "family activities", "children's activities", "playground", "kids",
    "suitable for kids", "fun for children", "games for kids", "family with kids", "kids area"
]

# Dropping duplicated columns and standardizing text
df6['reviews'] = df6['reviews'].str.strip().str.lower()
df6 = df6.drop_duplicates(subset=['name', 'formatted_address', 'reviews'])

# Define the function to analyze kids-friendly sentiment score
def analyze_kids_friendly_score(review_text):
    # Check if review_text is NaN and return default values if so
    if pd.isna(review_text):
        return 0, 0
    
    # Convert to string and split into sentences
    review_text = str(review_text)
    relevant_reviews = [sentence for sentence in review_text.split('.') if any(keyword in sentence for keyword in keywords_for_kids)]
    
    # If no relevant reviews are found, return default values
    if not relevant_reviews:
        return 0, 0
    
    # Analyze the sentiment of relevant sentences
    sentiment_scores = [TextBlob(sentence).sentiment.polarity for sentence in relevant_reviews]
    
    # Calculate the average sentiment score for kids-friendliness and the count of mentions
    avg_kids_friendly_sentiment = sum(sentiment_scores) / len(sentiment_scores)
    kids_mentions_count = len(relevant_reviews)
    
    return avg_kids_friendly_sentiment, kids_mentions_count

# Lists to store each place's sentiment score
kids_friendly_sentiments = []
kids_mentions_counts = []

# Iterate over each review to calculate the "kids-friendly" sentiment score
for index, row in df6.iterrows():
    review_text = row['reviews']  # Review text
    sentiment, mention_count = analyze_kids_friendly_score(review_text)
    kids_friendly_sentiments.append(sentiment)
    kids_mentions_counts.append(mention_count)

# Add the calculated scores to the DataFrame
df6['kids_friendly_sentiment'] = kids_friendly_sentiments
df6['kids_mentions_count'] = kids_mentions_counts

# Apply the function to the 'formatted_address' column and save to 'zip_code'
df6['zip_code'] = df6['formatted_address'].apply(extract_zip_code)
df6['zip_code'] = df6['zip_code'].astype(str)

# Generate unique_id using the first 4 characters of 'name', 'zip_code', and the row index
df6['unique_id'] = df6.apply(lambda row: f"{row['name'][:4].replace(' ', '')}_{row['zip_code']}_{row.name}", axis=1)
df6['zip_code'] = df6['zip_code'].replace({'75116': '75016'})

# Dropping unnecessary column
df6 = df6.drop(columns='arrondissement')
df6 = df6.dropna(subset=['reviews'], how='all')

# Save the results in CSV and Excel formats
df6.to_csv('Gral_df_Scores.csv', index=False)
df6.to_excel('Gral_df_Scores.xlsx', index=False)

print("Kids-friendly sentiment scores have been added and data saved.")
```

```python
# Function to detect keywords with negation handling
def is_kids_friendly(review):
    review = str(review).translate(str.maketrans('', '', string.punctuation)).lower()
    for keyword in keywords_for_kids:
        # Search for the keyword in the review
        match = re.search(r'\b' + keyword + r'\b', review)
        if match:
            # Check for negation words before the keyword (up to 3 words before)
            context = review[max(0, match.start() - 20):match.start()]
            if not any(neg_word in context for neg_word in negation_keywords):
                return True
    return False
    
# Key words kids-friendly
keywords_for_kids = ["kids menu", "children", "kid-friendly", "for children", "allergen-free dishes", "child-friendly", "family", "children's games", 
    "kids welcome", "family-oriented", "baby", "family", "family-friendly environment", "play area", "children's activity", "games", 
    "family with children", "kids", "kid", "children", "childrens","kids", "children", "family-friendly",
                     "high chair", "play area", "kids menu", "baby friendly","kids friendly","baby area"]
negation_keywords = ["not", "no", "never", "none", "didn't", "doesn't", "without"]
# New column 'kids_friendly' 
data['kids_friendly'] = data['reviews'].apply(is_kids_friendly)
data = data.rename(columns={'zip_code': 'zipcode'})
data['type']='Restaurant'
data['type_id'] = data.apply(lambda row: f"{row['type'][:2].replace(' ', '')}-{row['zipcode']}", axis=1)
data['df_id']='kffr3075'
data['df_unique_id']=data['df_id']+'_'+(data.index+1).astype(str)
data.to_csv("Gral_df_Scores.csv", index=False, encoding="utf-8",errors="replace")
```
